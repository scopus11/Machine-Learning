```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# membaca dataset
data = pd.read_csv("../data/day.csv")
print("Total data:", len(data))
data.head()
```

```python
# membagi data jadi 80% training dan 20% testing
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)
```

```python
# membagi 10% dari data training jadi validation
train_data, val_data = train_test_split(train_data, test_size=0.1, random_state=42)
```

```python
print("Jumlah data training:", len(train_data))
print("Jumlah data validation:", len(val_data))
print("Jumlah data testing:", len(test_data))

print("\nData Training (5 baris teratas):")
print(train_data.head())

print("\nData Validation (5 baris teratas):")
print(val_data.head())

print("\nData Testing (5 baris teratas):")
print(test_data.head())
```

```python
train_data.to_csv("../data/train_data.csv", index=False)
val_data.to_csv("../data/val_data.csv", index=False)
test_data.to_csv("../data/test_data.csv", index=False)
```
